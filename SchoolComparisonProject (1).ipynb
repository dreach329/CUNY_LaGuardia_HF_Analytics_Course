```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
dfBrooklyn = df[df['School DBN'].str.contains("K")]
```

```python
dfBrooklyn.shape
```

```python
dfBrooklyn.head(50)
```

```python
subset1 = dfBrooklyn["Mean Score"]
```

```python
print(subset1)
```

```python
subset1 = pd.to_numeric(subset1, errors="coerce")
```

```python
subset1.head()
```

```python
print(subset1.mean())
```

```python
dfONE = df[df['School Name'].str.contains("Midwood")]
```

```python
dfONE.head()
```

```python
subset2 = dfONE["Mean Score"]
```

```python
print(subset2)
```

```python
subset2 = pd.to_numeric(subset2, errors="coerce")
```

```python
print(subset2.mean())
```

```python
dfMAN = df[df['School DBN'].str.contains("M")]
```

```python
dfMAN.shape
```

```python
df.head(50)
```

```python
subset3 = dfMAN["Mean Score"]
```

```python
print(subset3)
```

```python
subset3 = pd.to_numeric(subset3, errors="coerce")
```

```python
subset3.head()
```

```python
print(subset3.mean())
```

```python
mean1 = subset1.mean()
mean2 = subset2.mean()
mean3 = subset3.mean()
```

```python
labels = ["Brooklyn", "Midwood High School", "Manhattan"]
means = [mean1, mean2, mean3]
```

```python
import matplotlib.pyplot as plt

plt.bar(labels, means)

plt.title("Average Scores for Each Subset")
plt.xlabel("Subsets")
plt.ylabel("Mean Score")

plt.show()
```

```python
Midwood High School Mean Score is higher than both Brooklyn and Manhattan average
```
